# dalex — Broad Model-Agnostic Explainability

DALEX (Descriptive mAchine Learning EXplanations) is a model-agnostic explainability framework originating from R, offering iBreakDown, PDP, and variable importance alongside Shapley value estimation. This notebook evaluates DALEX's Python port against `shapiq`'s Shapley interaction methods, comparing explanation faithfulness and runtime on the benchmark datasets. The focus is on how iBreakDown's local Shapley-based explanations compare to `shapiq`'s full interaction decomposition.
