# SAGE — Global feature importance

`SAGE` (Shapley Additive Global Explanations) computes global feature importance by applying Shapley values to the model's **loss function** rather than individual predictions, quantifying how much each feature contributes to overall model performance. Unlike local attribution methods, SAGE produces a single signed value per feature across the entire dataset, where negative values indicate features that actively harm predictions. This notebook benchmarks `SAGE` against `shapiq`'s mean absolute local Shapley values on the California Housing dataset to examine when these conceptually distinct global summaries agree and where they diverge. The comparison directly addresses Research Question X: how loss-based global Shapley values relate to prediction-based aggregated local Shapley values in terms of feature rankings, runtime, and stability.
