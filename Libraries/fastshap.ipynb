# fastshap — Amortized SHAP Approximation

FastSHAP learns a neural surrogate explainer that amortizes the cost of Shapley value computation, trading some accuracy for significant speed gains at inference time. This notebook benchmarks fastshap against `shapiq`'s exact and approximate Shapley methods, examining whether the surrogate approach can match marginal Shapley faithfulness at reduced runtime cost. The evaluation uses the Adult and Superconductivity datasets across varying feature counts.
