# shapleyflow — Graph-Based Shapley Path Decomposition

ShapleyFlow decomposes feature effects into direct and mediated (graph path) components, revealing high-order influence that standard Shapley interaction indices do not capture. This notebook explores shapleyflow's path-based decomposition on the benchmark datasets and contrasts its high-order interaction representation with `shapiq`'s SII, STI, and FSI indices. The evaluation addresses Research Question 5: whether path decomposition surfaces qualitatively different structure than existing interaction methods.
