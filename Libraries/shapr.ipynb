# shapr — Conditional Shapley Values

`shapr` computes conditional Shapley values — as opposed to marginal — accounting for feature correlations through a kernel-based conditional distribution estimator. This notebook benchmarks `shapr` against `shapiq` and `shap` on the high-correlation Superconductivity dataset, where conditional and marginal Shapley methods are expected to diverge most strongly. The comparison directly addresses Research Question 3: how conditional methods differ from marginal approaches on correlated data.
